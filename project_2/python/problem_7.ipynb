```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from analytical_eig import analytical_eigvec, analytical_eigval
```

```python
eigval = pd.read_csv("../Data/problem_7/Eigenvalues_9.csv",header=None)
eigvec = pd.read_csv("../Data/problem_7/Eigenvectors_9.csv",header=None)

min_ix = eigval.sort_values(0).iloc[:3].index

min_3 = eigvec.loc[:,min_ix]
min_3.index = np.linspace(1/10,9/10,9)
min_3.loc[0] = 0 
min_3.loc[1] = 0 
min_3.columns = eigval.loc[min_ix].squeeze().round(4)
min_3.sort_index().plot(figsize=(16,8))
plt.legend(title='Eigenvalues:')
plt.savefig("../Figures/eigenvectors_9.pdf")
```
