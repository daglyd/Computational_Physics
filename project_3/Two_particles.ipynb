```python
df = pd.read_csv('Data/main_two_particle_exp/positions_euler_cromer.csv',index_col=['Particle','x'])

x_start = df.loc[0].reset_index().iloc[0,0]
y_start = df.loc[0].reset_index().iloc[0,1]
x_end = df.loc[0].reset_index().iloc[-1,0]
y_end = df.loc[0].reset_index().iloc[-1,1]
z_start = df.loc[0].reset_index().iloc[0,2]
z_end = df.loc[0].reset_index().iloc[-1,2]

fig,ax = plt.subplots(1,2,figsize=(12,6))
ax[0].plot(df.loc[0,'y'],label='Particle 1')
ax[0].plot(x_start,y_start,'ro',label='start')
ax[0].plot(x_end,y_end,'go',label='end')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_xlim(-1000,1000)
ax[0].set_ylim(-1000,1000)
ax[0].legend()

x_start2 = df.loc[1].reset_index().iloc[0,0]
y_start2 = df.loc[1].reset_index().iloc[0,1]
x_end2 = df.loc[1].reset_index().iloc[-1,0]
y_end2 = df.loc[1].reset_index().iloc[-1,1]
z_start2 = df.loc[1].reset_index().iloc[0,2]
z_end2 = df.loc[1].reset_index().iloc[-1,2]

ax[1].plot(df.loc[1,'y'],label='Particle 2')
ax[1].plot(x_start2,y_start2,'ro',label='start')
ax[1].plot(x_end2,y_end2,'go',label='end')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_xlim(-1000,1000)
ax[1].set_ylim(-1000,1000)
ax[1].legend()
plt.savefig('Figures/main_two_particle_exp/xy_plot_euler_cromer.pdf')

fig,ax = plt.subplots(1,2,figsize=(12,6))
ax[0].plot(df.loc[0,'z'],label='Particle 1')
ax[0].plot(x_start,z_start,'ro',label='start')
ax[0].plot(x_end,z_end,'go',label='end')
ax[0].set_xlabel('x')
ax[0].set_ylabel('z')
ax[0].legend()
ax[1].plot(df.loc[1,'z'],label='Particle 2')
ax[1].plot(x_start2,z_start2,'ro',label='start')
ax[1].plot(x_end2,z_end2,'go',label='end')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
ax[1].legend()
plt.savefig('Figures/main_two_particle_exp/xz_plot_euler_cromer.pdf')
```

```python
combined = pd.concat([pd.read_csv('Data/main_two_particle_exp/velocities_euler_cromer.csv'),
           pd.read_csv('Data/main_two_particle_exp/positions_euler_cromer.csv').drop('Particle',axis=1)], axis=1)

x_start = combined.loc[combined.Particle==0].iloc[0,4]
vx_start = combined.loc[combined.Particle==0].iloc[0,1]
x_end = combined.loc[combined.Particle==0].iloc[-1,4]
vx_end = combined.loc[combined.Particle==0].iloc[-1,1]
x_start2 = combined.loc[combined.Particle==1].iloc[0,4]
vx_start2 = combined.loc[combined.Particle==1].iloc[0,1]
x_end2 = combined.loc[combined.Particle==1].iloc[-1,4]
vx_end2 = combined.loc[combined.Particle==1].iloc[-1,1]

combined.loc[combined.Particle==0].plot(x='x',y='vx',label='Phase plot particle 1')
plt.ylabel('$v_x$')
plt.plot(x_start,vx_start,'ro',label='start')
plt.plot(x_end,vx_end,'go',label='end')
plt.savefig('Figures/main_two_particle_exp/phase_plot_x_vx_euler_cromer1.pdf')

combined.loc[combined.Particle==1].plot(x='x',y='vx',subplots=True,label='Phase plot particle 2')
plt.ylabel('$v_x$')
plt.plot(x_start2,vx_start2,'ro',label='start')
plt.plot(x_end2,vx_end2,'go',label='end')
plt.savefig('Figures/main_two_particle_exp/phase_plot_x_vx_euler_cromer.pdf')

combined.loc[combined.Particle==0].plot(x='z',y='vz',subplots=True,label='Phase plot particle 1')
plt.ylabel('$v_z$')

plt.savefig('Figures/main_two_particle_exp/phase_plotz_vzeuler_cromer.pdf')
```
