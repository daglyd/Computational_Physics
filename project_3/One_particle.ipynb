```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
def make_plots(file_path):
    df = pd.read_csv(f'{file_path}',header=None).rename(columns = {0:'x',1:'y',2:'z'})

    df.plot(x='x',y='y')
    plt.plot(df.x.iloc[0],df.y.iloc[0],'ro',label='start')
    plt.plot(df.x.iloc[-1], df.y.iloc[-1],'go',label='end')
    plt.ylabel('y')
    plt.legend()
    plt.savefig('Figures/main_one_particle_exp/xy_plot_one_particle.pdf')
    
    df.plot(x='x',y='z')
    plt.plot(df.x.iloc[0],df.z.iloc[0],'ro',label='start')
    plt.plot(df.x.iloc[-1], df.z.iloc[-1],'go',label='end')
    plt.ylabel('z')
    plt.legend()
    plt.savefig('Figures/main_one_particle_exp/xz_plot_one_particle.pdf')
    
    df.plot(x='y',y='z')
    plt.plot(df.y.iloc[0],df.z.iloc[0],'ro',label='start')
    plt.plot(df.y.iloc[-1], df.z.iloc[-1],'go',label='end')
    plt.ylabel('z')
    plt.legend()
    plt.savefig('Figures/main_one_particle_exp/yz_plot_one_particle.pdf')
```

```python
make_plots('Data/one_particle/positions_euler_cromer.csv')
```

```python
df = pd.read_csv('Data/main_one_particle_exp/positions_euler_cromer.csv',index_col=['Particle','x'])

fig,ax = plt.subplots(1,figsize=(12,6))
ax.plot(df.loc[0,'y'])
plt.savefig('Figures/main_one_particle_exp/xy_plot.pdf')

fig,ax = plt.subplots(1,figsize=(12,6))
ax.plot(df.loc[0,'z'])
plt.savefig('Figures/main_one_particle_exp/xz_plot.pdf')
```

```python
df = pd.read_csv('Data/main_multi_particle_exp/positions_euler_cromer.csv',index_col=['Particle','x'])

fig,ax = plt.subplots(1,2,figsize=(12,6))
ax[0].plot(df.loc[0,'y'])
ax[1].plot(df.loc[1,'y'])
plt.savefig('Figures/main_two_particle_exp/xy_plot.pdf')

fig,ax = plt.subplots(1,2,figsize=(12,6))
ax[0].plot(df.loc[0,'z'])
ax[1].plot(df.loc[1,'z'])
plt.savefig('Figures/main_two_particle_exp/xz_plot.pdf')
```
